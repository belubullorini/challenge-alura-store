### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

print(ingreso_tienda)
print(ingreso_tienda2)
print(ingreso_tienda3)
print(ingreso_tienda4)
```

## Resumen:
Tienda con ingreso mas alto 'Tienda 1'  
Tienda con ingreso mas bajo 'Tienda 4'

###

# 2. Ventas por categoría

```python
import pandas as pd

resumen_categorias = pd.DataFrame({
    'Tienda 1': tienda.groupby('Categoría del Producto')['Producto'].count(),
    'Tienda 2': tienda2.groupby('Categoría del Producto')['Producto'].count(),
    'Tienda 3': tienda3.groupby('Categoría del Producto')['Producto'].count(),
    'Tienda 4': tienda4.groupby('Categoría del Producto')['Producto'].count()
})

resumen_categorias

```

```python
categorias_ordenadas = pd.DataFrame({
    'Tienda 1': resumen_categorias['Tienda 1'].sort_values(ascending=False).index.tolist(),
    'Tienda 2': resumen_categorias['Tienda 2'].sort_values(ascending=False).index.tolist(),
    'Tienda 3': resumen_categorias['Tienda 3'].sort_values(ascending=False).index.tolist(),
    'Tienda 4': resumen_categorias['Tienda 4'].sort_values(ascending=False).index.tolist()
})

categorias_ordenadas
```

## Resumen:
Las primeras tres categorias mas vendidas en las 4 tiendas son:
 1. Muebles
 2. Electronicos
 3. Juguetes


# 3. Calificación promedio de la tienda


```python
promedio_tienda1 = tienda['Calificación'].mean()
promedio_tienda2 = tienda2['Calificación'].mean()
promedio_tienda3 = tienda3['Calificación'].mean()
promedio_tienda4 = tienda4['Calificación'].mean()
```

```python
promedios_calificacion = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación Promedio': [promedio_tienda1, promedio_tienda2, promedio_tienda3, promedio_tienda4]
})

promedios_calificacion
```

```python
promedios_calificacion_ordenados = promedios_calificacion.sort_values(by='Calificación Promedio', ascending=False)
promedios_calificacion_ordenados
```

##Resumen:
Tienda 3 → mejor satisfacción del cliente ✅   
Tienda 1 → menor satisfacción del cliente ⚠️

# 4. Productos más y menos vendidos

```python
mas_vendidos = [
    tienda['Producto'].value_counts().idxmax(),
    tienda2['Producto'].value_counts().idxmax(),
    tienda3['Producto'].value_counts().idxmax(),
    tienda4['Producto'].value_counts().idxmax()
]

menos_vendidos = [
    tienda['Producto'].value_counts().idxmin(),
    tienda2['Producto'].value_counts().idxmin(),
    tienda3['Producto'].value_counts().idxmin(),
    tienda4['Producto'].value_counts().idxmin()
]

```

```python
import pandas as pd

resumen_productos = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Más Vendido': mas_vendidos,
    'Menos Vendido': menos_vendidos
})

resumen_productos
```

# 5. Envío promedio por tienda

```python
import pandas as pd

costo_envio = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Costo de Envío Promedio': [
        round(tienda['Costo de envío'].mean(), 2),
        round(tienda2['Costo de envío'].mean(), 2),
        round(tienda3['Costo de envío'].mean(), 2),
        round(tienda4['Costo de envío'].mean(), 2)
    ]
})

costo_envio
```

##Resumen:
**Tienda mas cara en envios: TIENDA 1**  
**Tienda mas barata en envios: TIENDA 4**

# 6. Resultados de los análisis en gráficos


```python
# Ingresos totales por tienda

import matplotlib.pyplot as plt

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [1150880400, 1116343500, 1098019600, 1038375700]

plt.figure(figsize=(8,5))
plt.bar(tiendas, ingresos, color='skyblue')
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingreso Total')
plt.show()
```

```python
# Costo de envío promedio por tienda

plt.plot(
    costo_envio['Tienda'],
    costo_envio['Costo de Envío Promedio'],
    marker='o'
)
plt.title('Costo de Envío Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo Promedio ($)')
plt.grid(True)
plt.show()
```

```python
#Calificación promedio por tienda

plt.figure(figsize=(7,7))
plt.pie(
    promedios_calificacion_ordenados['Calificación Promedio'],
    labels=promedios_calificacion_ordenados['Tienda'],
    autopct='%1.2f%%',
    startangle=90,
    colors=['#FF7043', '#FFA726', '#FFCC80', '#FFE0B2']
)
plt.title('Calificación Promedio por Tienda')
plt.show()

```
